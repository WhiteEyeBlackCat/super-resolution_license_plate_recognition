```python
import torch
from transformers import AutoProcessor, AutoModelForImageTextToText

path = "./Qwen3-VL/Qwen3-VL-8B-Instruct"
processor = AutoProcessor.from_pretrained(path, trust_remote_code=True, use_fast=False)
model = AutoModelForImageTextToText.from_pretrained(
    path, device_map="auto", torch_dtype="auto", trust_remote_code=True
)

# 列出所有名字含 vision / visual / encoder 的子模組
keys = []
for name, _ in model.named_modules():
    #if any(k in name.lower() for k in ["vision", "visual", "encoder"]):
        keys.append(name)

print("\n".join(keys[:200]))   # 先看前 200 個
```

```python
from PIL import Image

img = Image.open("./challenge_development_set_final/plate/seq_000/000000.png").convert("RGB")

stash = {}

def hook_fn(m, inp, out):
    # inp 通常是 tuple
    print("[VISION] #inputs:", len(inp))
    for i, x in enumerate(inp):
        if hasattr(x, "shape"):
            print(f"[VISION] inp[{i}].shape:", tuple(x.shape), x.dtype, x.device)
        else:
            print(f"[VISION] inp[{i}] type:", type(x))

    y = out
    if hasattr(out, "last_hidden_state"):
        y = out.last_hidden_state
    if hasattr(y, "shape"):
        print("[VISION] out.shape:", tuple(y.shape), y.dtype, y.device)

    stash["inp"] = inp
    stash["out"] = out

h = vision.register_forward_hook(hook_fn)

# 走一次完整 forward（讓模型自己處理圖像路徑/佔位符）
messages = [{
  "role": "user",
  "content": [
    {"type": "image", "image": img},
    {"type": "text", "text": "只輸出 UNKNOWN"}
  ]
}]
text = processor.apply_chat_template(messages, add_generation_prompt=True)
inputs = processor(text=text, images=img, return_tensors="pt")
inputs = {k: v.to(model.device) for k, v in inputs.items()}

with torch.inference_mode():
    _ = model(**inputs)

h.remove()

out = stash["out"]
if hasattr(out, "last_hidden_state"):
    out = out.last_hidden_state
print("visual out shape:", getattr(out, "shape", None))
```
