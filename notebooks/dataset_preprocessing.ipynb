```python
import json
import grab
import os
from PIL import Image, ImageDraw
from IPython.display import display
```

```python
path = r"./challenge_development_set_final"

bounding_box = {}
for i in range(0,39):
    if i < 10:
        with open(f"./challenge_development_set_final/seq_00{i}/detections.json", "r", encoding="utf-8") as f:
            bounding_box[f"seq{i}"] = json.load(f)
    else:
        with open(f"./challenge_development_set_final/seq_0{i}/detections.json", "r", encoding="utf-8") as f:
            bounding_box[f"seq{i}"] = json.load(f)
```

```python
print(bounding_box['seq38'])
print(bounding_box['seq1'][0])
print(bounding_box['seq1'][0]['frame'])
```

```python
os.makedirs("plate", exist_ok=True)
for i in range(0,39):
    os.makedirs(f"plate/seq_{i:03d}", exist_ok=True)
    for j in range(0,10):
        img = Image.open(path+f"/seq_{i:03d}/0000{j:02d}.png")
        draw = ImageDraw.Draw(img)
        x1, y1, x2, y2 = bounding_box[f"seq{i}"][j]['license_plate_coordinates']
        bbox = (x1, y1, x2, y2)
        roi = img.crop(bbox)
        roi.save(f"plate/seq_{i:03d}/0000{j:02d}.png")

display(roi)
print(roi.size)
```
