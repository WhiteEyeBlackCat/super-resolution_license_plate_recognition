```python
import torch
import matplotlib.pyplot as plt
import numpy as np
import sys
import os
project_new_path = os.path.abspath("/home/re6141011/challenge/plate_recognition/new")
if project_new_path not in sys.path:
    sys.path.append(project_new_path)

from challenge.plate_recognition.new.model2 import LPLLM  


model_path = "/home/re6141011/challenge/plate_recognition/Qwen3-VL/Qwen3-VL-8B-Instruct"
# ========= 1. 建立模型 =========
model = LPLLM(
        model_path=model_path,
        num_slots=7,
        cmrm_dim=4096,
        num_heads=8,
    )

# ========= 2. 讀取 checkpoint =========
ckpt_path = "./new/checkpoint/best.pt"
ckpt = torch.load(ckpt_path, map_location="cpu")

print("Checkpoint keys:", ckpt.keys())
print("step:", ckpt["step"])
print("epoch:", ckpt["epoch"])
print("val_loss:", ckpt["val_loss"])

model.load_state_dict(ckpt["cmrm_state_dict"], strict=True)
model.eval()

# ========= 3. 收集所有權重 =========
all_weights = []

for name, param in model.named_parameters():
    if param.requires_grad and param.data.numel() > 0:
        all_weights.append(param.data.detach().cpu().numpy().ravel())

all_weights = np.concatenate(all_weights)

# ========= 4. 畫整體權重 histogram =========
plt.figure(figsize=(8, 5))
plt.hist(all_weights, bins=100)
plt.title("CMRM Weight Distribution (Today Checkpoint)")
plt.xlabel("Weight Value")
plt.ylabel("Count")
plt.grid(True, alpha=0.3)
plt.show()
```
